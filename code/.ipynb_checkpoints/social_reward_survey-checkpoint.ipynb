## Clean Data

```python
# Find the redcap exported CSVs and take the last one
redcap_file = glob.glob(rdcp_dir+'SCONNChildPacket-SocialRewardFMRIQues_DATA_*.csv')[-1]

# Read the data
redcap_orig = pd.read_csv(redcap_file)

# Make a copy of the original data
redcap_data = redcap_orig.copy()

# Make all IDs lowercase (for consistency)
redcap_data['record_id'] = redcap_data['record_id'].str.lower()

redcap_data = redcap_data[redcap_data['record_id'].str.startswith('scn')]

# Make the record id the index
redcap_data = redcap_data.set_index('record_id')

redcap_data
```

```python
# Drop irrelevant columns
relv_cols = [x for x in redcap_data.columns if x.startswith('socrew')]

redcap_data = redcap_data[relv_cols]

# Drop missing data
redcap_data = redcap_data.dropna()
redcap_data
```

# Similarity Analysis
Similarity is a measure of how "similar" two participants are. Dissimilarity is just the reverse of that, and so the two terms can be used interchangible, depending on how you want to describe the data. For example, people are similar in their interest (closer together in interest). But if you wanted to describe people in physical distance, it is helpful to think dissimilarity (people are 10 feet away, not 10 feet close). Use similarity/dissimilarity in whatever frame is most convinient, but be consistent for easier understanding. 

## Calculate dissimilarity
Here, I am using euclidean distance as a measure of dissimilarity. [Euclidean distance](https://en.wikipedia.org/wiki/Euclidean_distance) can be thought of as the distance between two points in some space. Just like how you had to find the distance between two points on a graph with X and Y coordinates. Except here there are not just two dimensions for each data point, but dimensions equal to the number of questions in the survey. Each participant's data can be described as the answers to (socrew1, socrew2, socrew2, ..., socrew211) in this high dimensional space, and then the distance between these data points can be measured. 

```python
from scipy.spatial.distance import euclidean, pdist, squareform
import matplotlib.pyplot as plt

# Create similarity matrix
dists = pdist(redcap_data)
similarity_matrix = pd.DataFrame(squareform(dists), 
                                 columns=redcap_data.index, index=redcap_data.index)


fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(similarity_matrix, cmap='YlGnBu', linewidths=.5, ax=ax)
```

That is what is shown in the heatmap above. The darker colors represent larger distances (more dissimilar) between particpants, while the lighter colors represent particpants that are closer (more similar).
